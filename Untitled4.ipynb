```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_regression
```

```python
data1 = make_regression(n_samples=50, n_features=6, n_informative=6, n_targets=1, 
                        bias=0.0, effective_rank=None,tail_strength=0.5, 
                        noise=0.0, shuffle=True, coef=False, random_state=None)
df1 = pd.DataFrame(data1[0],columns=['x'+str(i) for i in range(1,7)])
df1['y'] = data1[1]
```

```python
X=res[['N','P','Ca','K','Na','Mg']]
y=res['pH']
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test =train_test_split(X , y , test_size=0.2 , random_state=4)
```

```python
print('Train set :', X_train.shape , y_train.shape)
print('Test set :' , X_test.shape , y_test.shape)
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn import linear_model
```

```python
poly=PolynomialFeatures(1)
train_X_poly=poly.fit_transform(X_train)
train_X_poly
```

```python
df=linear_model.LinearRegression()
train_y=df.fit(train_X_poly , y_train)
```

```python
test_X_poly=poly.fit_transform(X_test)
test_X_poly
```

```python
test_hat=df.predict(test_X_poly)
```

```python
from sklearn.metrics import r2_score
print("R2-score :", r2_score(test_hat , y_test))
```
